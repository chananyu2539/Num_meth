# 1D Gradient Descent

1) Let us consider the function
$$ f(x) = (x-2)^2 + 1 $$
First plot it.

```python
def f(x):
    return (x-2)**2+1

def df(x):
    return 2*(x-2)

xs = np.linspace(0,5,100)
plt.plot(xs,[f(x) for x in xs])
```

2) If we start at $x=4$, which way should we move and why?

negative, minimum at 2, slope at 4 is positive

3)If the step size is constants = 1.1, what does the update rule look like?

$$x_{n+1} = \ldots$$

$$ x_{n+1} = x_n + (-\frac{slope}{abs(slope)}) * stepsize $$

4)Let try it with step size = 1.1. For 10 iteration. Print and plot your point as you go. Can you understand the behavior?

```python
cx = 4.
stepsize = 1.1
testx = []
testy = []
for i in range(10):
    testx.append(cx)
    testy.append(f(cx))
    cx = cx +(-df(cx)/np.abs(df(cx)))*stepsize

plt.plot(xs,[f(x) for x in xs])
plt.plot(testx,testy,'o-')
plt.grid(True)
print testx
```

5)Now let's try step size=0.01 for 100 iteration. Print and plot your point as you go. Can you understand the behavior?

```python
cx = 4.
stepsize = 0.01
testx = []
testy = []
for i in range(100):
    testx.append(cx)
    testy.append(f(cx))
    cx = cx +(-df(cx)/np.abs(df(cx)))*stepsize

plt.plot(xs,[f(x) for x in xs])
plt.plot(testx,testy,'o-')
plt.grid(True)
# print testx
```

6) When do we want the step size to be small and when do you want the step size to be big?

Big: when we are far from answer,  Small: when close

7) What function kind of does that?

slope

8)If we use that function, what does the update rule looks like.

$$ x_{n+1} = x_n - lamda * slope $$

9) Try it with $\lambda=0.1$. Print and Plot to see what happens?

```python
cx = 4.
ld = 0.1
testx = []
testy = []
for i in range(100):
    testx.append(cx)
    testy.append(f(cx))
    cx = cx - ld*df(cx)

plt.plot(xs,[f(x) for x in xs])
plt.plot(testx,testy,'o-')
plt.grid(True)
```

10) Try it with $\lambda = 1.1$. For 4 iterations. Can you see how it goes?

```python
cx = 4.
ld = 1.1
testx = []
testy = []
for i in range(4):
    testx.append(cx)
    testy.append(f(cx))
    cx = cx - ld*df(cx)

plt.plot(xs,[f(x) for x in xs])
plt.plot(testx,testy,'o-')
plt.grid(True)
```

# Multi-Dimension Gradient Descent

1) Let us consider the function.
$$f(x,y) = (x-2)^2 + xy + 2y + 1$$
Make the 3D plot.

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
#see matplotlib3d surface
def f(x,y):
    return (x-2)**2+x*y+y**2+1
nx = 200
ny = 200
x = np.linspace(-20,20,nx)
y = np.linspace(-20,20,ny)
xgrid, ygrid = np.meshgrid(x, y)
#print ygrid
uz = np.zeros((nx,ny))
for ix in range(nx):
    for iy in range(ny):
        xx = x[ix]
        yy = y[iy]
        uz[iy,ix] = f(xx,yy) ##<<<<<<This is not a typo: [iy,ix]
        
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.gca().view_init(elev=15., azim=30)
ucs = ax.plot_surface(xgrid, ygrid, uz, cmap=cm.coolwarm)
plt.title('f(x,y)')
```

###Linear Regression Revisited

Let us consider the problem of findind a line that pass through the points.

```python
np.random.seed(123)
n = 100
x = np.linspace(0,1,n)
y = -2*x + 1 + np.random.randn(n)/2
plt.plot(x,y,'.r')
```

1) What exactly are we trying to minimize and why?

```python
def cost(mc):
    m,c = mc
    return sum((m*x+c-y)**2)


def gradf(mc):
    h = 1e-3
    return np.array([(cost(mc+[h,0])-cost(mc-[h,0]))/(2*h),(cost(mc+[0,h])-cost(mc-[0,h]))/(2*h)])
    


def gradd(v,l=0.1,n=10000):
#     plt.plot(v[0],v[1],'ob')
    for i in range(n):
#         print gradf(v)
        v = v - l*gradf(v)
#         plt.plot(v[0],v[1],'ob')
    return v
        
# print cost(np.array([2.,2.]))

v = gradd(np.array([0.,1.]),0.001)
plt.plot(x,y,'.r')
plt.plot(x,[v[0]*xx+v[1] for xx in x])
        
    
```

```python
np.random.seed(123)
n = 100
x = np.linspace(0,1,n)
y = -2*x + 1 + np.random.randn(n)/2
plt.plot(x,y,'.r')

import scipy.optimize

res = scipy.optimize.minimize(cost,[-1.,2.])
m,c = res.x
plt.plot(x,m*x+c)
```

# Separating Line.

We want to draw a separating line betwen two datasets shown below.

```python
np.random.seed(234)
n1 = 100
n2 = 100
mean = [0, 0]
cov = [[1, -0.5], [-0.5, 1]]
r1, r2 = np.random.multivariate_normal(mean, cov, n1).T
#plt.plot(r1, r2, '.r')

mean = [2, 2]
cov = [[1, 0], [0, 1]]
b1, b2 = np.random.multivariate_normal(mean, cov, n2).T
#plt.plot(b1, b2, '.b')

c = np.zeros(n1+n2)
c[:n1] = 1.
c[n1:] = 0.
#print c

d = np.zeros((n1+n2,2))
d[:n1,0] = r1
d[:n1,1] = r2
d[n1:,0] = b1
d[n1:,1] = b2
```

```python
plt.plot(d[c==1,0], d[c==1,1],'.r')#note boolean indexing
plt.plot(d[c==0,0], d[c==0,1],'.b')#note boolean indexing
```
