1.1) Use gaussian elimination to solve the following system of equations. You can do it by hands. No need to implement one.

$$
\begin{bmatrix}
-1 & 2 & 1 \\
2 & 3 & 4 \\
10 & -5 & 2
\end{bmatrix}
\begin{bmatrix}
x\\
y \\
z
\end{bmatrix}
=
\begin{bmatrix}
2 \\
6 \\
22
\end{bmatrix}
$$

```python
def forward(A,C):
    Ai = A.copy()
    Ci = C.copy()
    row = Ai.shape[0]
    for i in range(row):
        for j in range(i+1,row):
            Ci[j] = Ci[j] - Ai[j,i]/Ai[i,i]*Ci[i]
            Ai[j,:] = Ai[j,:] - Ai[j,i]/Ai[i,i]*Ai[i,:]
    return Ai,Ci

def backward(A,C):
    Ai = A.copy()
    Ci = C.copy()
    ans = np.zeros(Ci.shape)
    for i in range(Ai.shape[0]-1,-1,-1):
        csum = 0
        for j in range(i+1,Ai.shape[0]):
#             print i,j
            csum += Ai[i,j]*ans[j]     
        ans[i] = (Ci[i] - csum)/Ai[i,i]
    return ans

A = np.array([[-1.,2.,1.],
              [2.,3.,4.],
              [10.,-5.,2.]])
B = np.array([2.,6.,22.])
    
    
nA,nC = forward(A,B)  
aAns = backward(nA,nC)

print aAns
```

#### 2) System of Non-linear Equations

Find a solution to the following system of non linear equations using Newton's Method.
$$x^2 + y^2 = 26.9165$$
$$xy + x^2 = 22.8278$$

```python
def u(x,y):
    return x**2+y**2 - 26.9165

def v(x,y):
    return x*y + x**2 - 22.8278

```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

nx = 200
ny = 200
x = np.linspace(-20,20,nx)
y = np.linspace(-20,20,ny)
xgrid, ygrid = np.meshgrid(x, y)
#print ygrid
uz = np.zeros((nx,ny))
vz = np.zeros((nx,ny))
for ix in range(nx):
    for iy in range(ny):
        xx = x[ix]
        yy = y[iy]
        uz[iy,ix] = u(xx,yy) ##<<<<<<This is not a typo: [iy,ix]
        vz[iy,ix] = v(xx,yy)


ucs = plt.contour(x, y, uz, colors=('r'))
plt.clabel(ucs, inline=1, fontsize=10)
vcs = plt.contour(x, y, vz, colors=('b'))
plt.clabel(vcs, inline=1, fontsize=10)
plt.grid()
```

```python
# -3,-3
# -5,2
# 4,0
# 3,4
def u(x,y):
    return x**2+y**2 - 26.9165

def v(x,y):
    return x*y + x**2 - 22.8278

def uX(x,y):
    return 2.*x
def uY(x,y):
    return 2.*y

def vX(x,y):
    return 2.*x + y
def vY(x,y):
    return x
```

```python
# -3,-3
# -5,2
# 4,0
# 3,4

def newton2d(u,v,ux,uy,vx,vy,x0,y0):
    for i in range(3):
        C = np.array([-u(x0,y0)+ux(x0,y0)*x0+uy(x0,y0)*y0,-v(x0,y0)+vx(x0,y0)*x0+vy(x0,y0)*y0])
        A = np.array([[ux(x0,y0),uy(x0,y0)],
                      [vx(x0,y0),vy(x0,y0)]])
        fA,fC = forward(A,C)
        ans = backward(fA,fC)
        x0,y0 = ans[0],ans[1]
    
    return x0,y0
  
x1,y1 =  newton2d(u,v,uX,uY,vX,vY,-3.,-3.)
print x1,y1,u(x1,y1)

x2,y2 =  newton2d(u,v,uX,uY,vX,vY,-5.,2.)
print x2,y2,u(x2,y2)

x3,y3 =  newton2d(u,v,uX,uY,vX,vY,4.,0.)
print x3,y3,u(x3,y3)

x4,y4 =  newton2d(u,v,uX,uY,vX,vY,3.,4.)
print x4,y4,u(x4,y4)


print u(x1,y1),v(x1,y1),u(x2,y2),v(x2,y2),u(x3,y3),v(x3,y3),u(x4,y4),v(x4,y4)
```
